### Exercise 1 - Turn the shopping cart program from yesterday into an object-oriented program
The comments in the cell below are there as a guide for thinking about the problem. However, if you feel a different way is best for you and your own thought process, please do what feels best for you by all means.

```python
class Cart:
    while True:
        ask = input("Would you like to add/remove/ or quit and show your shopping cart?")
        break
    def __init__ (self):
        self.cart = []
    def add(self):
        shopping = input('What would you like to add to your cart?')
        if shopping not in self.cart:
            self.cart.append(shopping)
            print(self.cart) 
    def remove(self):
        shopping = input('What would you like to remove from your cart?')
        if shopping in self.cart:
            self.cart.remove(shopping)
            print(f'{shopping} has been removed from your cart')
    def quit(self):
        shopping = input('Do you want to quit and show cart? Yes or No?').lower()
        if shopping == "Yes":
            print(self.cart)
        else:
            self.add()
            
        
my_cart = Cart()

       

my_cart.add()
my_cart.remove()
my_cart.quit()

Cart()
```
